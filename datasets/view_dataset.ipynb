```python
import pandas as pd
```

```python
dataset = "newsgroups"
dataset = "spotify"
dataset = "amazon"
```

```python
dataset2trainpath = {
    "spotify": "datasets/spotify/data/spotify_train.parquet",
    "newsgroups": "datasets/newsgroups/data/newsgroups_train.parquet",
    "amazon": "datasets/amazon/data/amazon_train.parquet"
}

dataset2testpath = {
    "spotify": "datasets/spotify/data/spotify_test.parquet",
    "newsgroups": "datasets/newsgroups/data/newsgroups_test.parquet",
    "amazon": "datasets/amazon/data/amazon_test.parquet"
}
```

```python
df = pd.read_parquet(dataset2testpath[dataset])
df.head(10)
```

```python
# remove rows where text is empty (len=0) or NaN
empty_rows = df[(df["text"].isna()) | (df["text"].str.len() == 0)]
print(f"Number of empty rows in {dataset} test set: {len(empty_rows)}")

# remove empty rows
df = df[~((df["text"].isna()) | (df["text"].str.len() == 0))]
```

```python
# rename columns product-summary -> product_category-summary
df.rename(columns={"label": "topic"}, inplace=True)
```

```python
# rename columns product-summary -> product_category-summary
df.rename(columns={"product-summary": "product_category-summary"}, inplace=True)
df.rename(columns={"product-keyphrases": "product_category-keyphrases"}, inplace=True)
df.rename(columns={"sentiment-summary": "stars-summary"}, inplace=True)
df.rename(columns={"sentiment-keyphrases": "stars-keyphrases"}, inplace=True)
```

```python
df.to_parquet(dataset2trainpath[dataset], index=False)
```
